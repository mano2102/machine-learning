#Creating a rank 4 tensor

```python
rank_4_tensor=tf.zeros(shape=(2,3,4,5))
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
# Printing all the tensor attributes

print("Dimensions of the tensor:",rank_4_tensor.ndim)
print("Shape of the tensors:",rank_4_tensor.shape)
print("Data type of the tensors",rank_4_tensor.dtype)
print("Element along the 0 axis",rank_4_tensor[0])
print("Element along the last axis",rank_4_tensor[-1])
print("Total number of elements in the tensor using tf:",tf.size(rank_4_tensor))
print("Total number of elements in the tensor using numpy:",tf.size(rank_4_tensor).numpy())
```
